```python
import time
committees = load_committees_in_period(ticker='BTC_USDT', exchange='binance', start_timestamp=0, end_timestamp=time.time())
```

```python
def backtest(doge_trader, start_time, end_time, ticker, exchange, horizon=12):
    from apps.genetic_algorithms.gp_data import Data
    data = Data(start_time, end_time, ticker, horizon, start_cash=1000, start_crypto=0, source=exchange)
    evaluation = doge_trader.gp.build_evaluation_object(doge_trader.doge, data)
    return evaluation
    
```

```python
from apps.genetic_algorithms.chart_plotter import get_dot_graph
from IPython.display import display
from apps.backtesting.data_sources import db_interface
ticker = 'BTC_USDT'

end = db_interface.get_nearest_db_timestamp(time.time(), ticker, exchange="binance")
start = db_interface.get_nearest_db_timestamp(end-60*60*24, ticker, exchange="binance")
print(start, end)


for committee in committees:
    for trader in committee.doge_traders:
        display(get_dot_graph(trader.doge))
        try:
            evaluation = backtest(trader, end_time=end, start_time=start, ticker='BTC_USDT', exchange='binance')
            print(evaluation.get_report())
        except Exception as e:
            print(e)
        
        
        
```
